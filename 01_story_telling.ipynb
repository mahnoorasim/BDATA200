<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Sumo Wrestlers: Honor... or Cheating?</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>When Incentives Override Tradition</em></p></div>### **THE SETUP**- Sumo is Japan's most sacred sport, built on centuries of honor- Wrestlers with **7-7 records** on the final day desperately need 1 more win to keep their ranking- Their opponent at **8-6** has nothing to lose### **WHAT THE DATA SHOWS**- The 7-7 wrestler wins **~75%** of the time (expected: ~50%)- In rematches, the pattern mysteriously **reverses**- When media investigated, the suspicious wins suddenly **disappeared**> **Data lesson:** Incentives shape behavior — even in the most honor-bound traditions.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why Did Crime Drop in the 1990s?</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>The Most Controversial Finding in Freakonomics</em></p></div>### **EVERYONE SAID IT WAS BECAUSE OF:**- Better policing strategies- Harsher prison sentences- A booming economy### **LEVITT'S CONTROVERSIAL FINDING**- The biggest factor was the legalization of abortion (*Roe v. Wade*, 1973)- Unwanted children born in high-risk environments are more likely to become criminals- 18 years after 1973, those never-born children would have been peak crime age- States that legalized abortion earlier saw crime drop earlier> **Data lesson:** The real cause is often hidden — and controversial.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Do Real Estate Agents Work for YOU?</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>The Information Asymmetry Problem</em></p></div>### **THE PROBLEM**- Your agent earns ~1.5% commission on your home sale- On a \K house, an extra \K for you = only **\ more** for them- So their incentive is to sell **fast**, not sell **high**### **THE PROOF**- When agents sell **their own homes**, they keep them on market **10 extra days**- And sell for **3% more** than comparable client homes!> **Data lesson:** Incentives reveal true behavior.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Teachers Cheating? Yes, Really!</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>When High Stakes Create Perverse Incentives</em></p></div>### **CHICAGO PUBLIC SCHOOLS STANDARDIZED TESTING**- High-stakes testing: schools get rewarded or punished based on scores- Teachers had access to answer sheets before submitting them### **HOW DATA CAUGHT THEM**- Levitt's algorithm found suspicious answer patterns: identical strings of correct answers, especially on hard questions- Classrooms with dramatic score jumps followed by drops the next year- Result: **~5% of teachers were cheating**. Many were fired!> **Data lesson:** Algorithms can detect fraud humans miss.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">What Makes a Perfect Parent?</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>What the Data Says vs. What We Believe</em></p></div><table style="width: 100%; border-collapse: collapse; margin: 20px 0; box-shadow: 0 2px 8px rgba(0,0,0,0.1);">  <tr>    <th style="padding: 12px 15px; border: 1px solid #4B2E83; background-color: #4B2E83; color: white; text-align: left;">What DOES Matter (Data Says)</th>    <th style="padding: 12px 15px; border: 1px solid #4B2E83; background-color: #4B2E83; color: white; text-align: left;">What DOESN'T Matter (Surprise!)</th>  </tr>  <tr>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5;">Parents are highly educated</td>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5;">Family is "intact" (two-parent household)</td>  </tr>  <tr>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5; background-color: #F9F7FC;">Parents have high socioeconomic status</td>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5; background-color: #F9F7FC;">Parents read to child every day</td>  </tr>  <tr>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5;">Mom was 30+ at birth of first child</td>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5;">Child regularly visits museums</td>  </tr>  <tr>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5; background-color: #F9F7FC;">Child has many books in the home</td>    <td style="padding: 10px 15px; border: 1px solid #E8E0F5; background-color: #F9F7FC;">Child watches little TV</td>  </tr></table>> **Data lesson:** Who the parents **are** matters more than what they **do**.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Does Your Name Shape Your Destiny?</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>A Tale of Two Brothers</em></p></div>### **THE STORY**- A mother named her sons **"Winner"** and **"Loser"**- Winner became a criminal. Loser became a detective!### **WHAT THE DATA ACTUALLY SHOWS**- Names don't **cause** success or failure- But names **signal** the parents' education and economic status- Names trickle down the economic ladder over time- "Distinctly Black" names had **50% lower** callback rates on resumes (audit study) — revealing **prejudice**, not destiny> **Data lesson:** Correlation is not causation!

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Why Do Drug Dealers Live with Their Moms?</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>When Real Data Destroys Hollywood Myths</em></p></div>### **THE MYTH**- Drug dealing is hugely profitable### **THE DATA (from actual gang financial records!)**- A sociologist named Sudhir Venkatesh got access to a gang's books- The gang was structured like McDonald's — a **franchise model**!- Top boss: **\K/year**. Average foot soldier: **\.30/hour**- That's **below minimum wage** — hence living with Mom- **1-in-4 chance** of being killed in 4 years on the job> **Data lesson:** Real data destroys Hollywood myths.

<div style="border-left: 5px solid #4B2E83; background-color: #F9F7FC; padding: 20px 25px; margin: 20px 0; border-radius: 5px;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">The Art of Data Storytelling</h2>  <p style="font-size: 1.05em; color: #555; line-height: 1.7;"><em>What Makes Freakonomics a Storytelling Masterclass</em></p></div>- **Start with a question, not an answer** — "Why do drug dealers live with their moms?" hooks attention instantly- **Challenge assumptions with data** — the sumo & teacher cheating stories show data overturning "common sense"- **Use surprise as your weapon** — unexpected findings (\.30/hr for drug dealers) are unforgettable- **Make data human** — every chart tells a story about real people making real choices- **Build a narrative arc** — setup &rarr; data reveal &rarr; insight keeps audiences engaged from start to finish<div style="background-color: #F9F7FC; border: 2px solid #4B2E83; padding: 15px 20px; border-radius: 8px; margin: 20px 0; text-align: center;">  <p style="font-size: 1.15em; color: #4B2E83; font-weight: 600; margin: 0;">Great data storytelling = the right question + surprising data + a human narrative</p></div>

<div style="background: linear-gradient(135deg, #E8D3FF 0%, #F9F7FC 100%); padding: 25px; border-radius: 10px; border: 2px solid #4B2E83; margin: 30px 0;">  <h2 style="color: #4B2E83; font-family: 'Encode Sans', 'Arial', sans-serif; margin-top: 0;">Key Takeaways - Part 4</h2>  <ul style="font-size: 1.05em; line-height: 1.8; color: #333;">    <li><strong>Incentives drive behavior:</strong> sumo wrestlers, teachers, and real estate agents all respond to incentives</li>    <li><strong>Correlation is not causation:</strong> names don't cause success; churches don't cause crime</li>    <li><strong>Data destroys myths:</strong> real evidence often contradicts "common sense"</li>    <li><strong>Algorithms detect patterns:</strong> data analysis can catch fraud that humans miss</li>  </ul></div>
